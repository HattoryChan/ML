### Euclidean distance

```python
from tqdm import tqdm_notebook as tqdm
```

```python
distance_min =10e23
index = 3
for row_x in tqdm(df_Data_X.to_numpy(), desc = "Rows X: "):
    for row_y in df_Data_Clear[df_Data_Clear.msisdn == df_Facts.iloc[index].Second].to_numpy():
        
        dist = distance.euclidean(row_x, row_y)
        if (dist < distance_min):
            distance_min = dist 
            row_x_similar_y = row_x       
```

```python
print("Msisdn: ", df_Facts.iloc[index].Second)
print("Right msisdn: ", df_Facts.iloc[index].First)
print("Answer msisdn: ", pd.Series(row_x_similar_y).astype('int')[2] )
print("Answer distance: ", distance_min)
```

### MAE

```python
from sklearn.metrics import mean_absolute_error
```

```python
distance_min =10e23

for row_x in tqdm(df_Data_X.to_numpy(), desc = "Rows X: "):
    for row_y in df_Data_Clear[df_Data_Clear.msisdn == df_Facts.iloc[index].Second].to_numpy():
        
        dist = mean_absolute_error(row_x, row_y)
        if (dist < distance_min):
            distance_min = dist 
            row_x_similar_y = row_x  
```

```python
print("Msisdn: ", df_Facts.iloc[index].Second)
print("Right msisdn: ", df_Facts.iloc[index].First)
print("Answer msisdn: ", pd.Series(row_x_similar_y).astype('int')[2] )
print("Answer distance: ", distance_min)
```

### Cosine

```python
from scipy.spatial import distance
```

```python
distance_min =10e23

for row_x in tqdm(df_Data_X.to_numpy(), desc = "Rows X: "):
    for row_y in df_Data_Clear[df_Data_Clear.msisdn == df_Facts.iloc[index].Second].to_numpy():
        
        dist = distance.cosine(row_x, row_y)
        if (dist < distance_min):
            distance_min = dist 
            row_x_similar_y = row_x  
```

```python
print("Msisdn: ", df_Facts.iloc[index].Second)
print("Right msisdn: ", df_Facts.iloc[index].First)
print("Answer msisdn: ", pd.Series(row_x_similar_y).astype('int')[2] )
print("Answer distance: ", distance_min)
```

### Chebyshev

```python
distance_min =10e23

for row_x in tqdm(df_Data_X.to_numpy(), desc = "Rows X: "):
    for row_y in df_Data_Clear[df_Data_Clear.msisdn == df_Facts.iloc[index].Second].to_numpy():
        
        dist = distance.chebyshev(row_x, row_y)
        if (dist < distance_min):
            distance_min = dist 
            row_x_similar_y = row_x  
```

### Calculate STD

```python
df_Data_Std = pd.DataFrame(columns= df_Data.columns)
for i in tqdm(range(len(set(df_Facts.First)))):
    df_Data_Std = pd.concat([df_Data_Std, 
                              df_Data[df_Data.msisdn == df_Facts.iloc[i].First].std().to_frame().T], axis=0)
    df_Data_Std.iloc[i].msisdn = df_Data[df_Data.msisdn == df_Facts.iloc[i].First].mean().msisdn
    
for i in tqdm(range(len(set(df_Facts.Second)))):
    df_Data_Std = pd.concat([df_Data_Std, 
                              df_Data[df_Data.msisdn == df_Facts.iloc[i].Second].std().to_frame().T], axis=0)
    df_Data_Std.iloc[i+120].msisdn = df_Data[df_Data.msisdn == df_Facts.iloc[i].Second].mean().msisdn
   
```

### Simple Tree

```python
import xgboost as xgb
from sklearn.metrics import accuracy_score
```

```python
df_Data_Clear.head()
```

```python
df_Facts.shape
```

```python
df_Data_Clear.shape
```

```python
df_Data_Train = pd.concat([df_Data_X, df_Data_Y], axis=0).reset_index()
```

```python
df_Data_Train_Y = pd.DataFrame(columns= ['msisdn'])
for i in tqdm(range(len(set(df_Data_X.index)))):
    df_Data_Train_Y = pd.concat([df_Data_Train_Y, 
                                 df_Facts[df_Facts.First == df_Data_X.iloc[i].msisdn].Second.head(1)],
                                axis=0, ignore_index=True)
    
for i in tqdm(range(len(set(df_Data_Y.index)))):
    df_Data_Train_Y = pd.concat([df_Data_Train_Y, 
                                 df_Facts[df_Facts.Second == df_Data_Y.iloc[i].msisdn].First.head(1)],
                               axis=0,  ignore_index=True) 

df_Data_Train_Y = df_Data_Train_Y.drop(['msisdn'], axis=1)    
df_Data_Train_Y.columns = ['msisdn']    
```

```python
df_Data_Train_Y.head()
```

```python
df_Data_Train_Y.shape, df_Data_Train.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_Data_Train, df_Data_Train_Y,
                                                   test_size = 0.2,
                                                   random_state = 17)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(X_train, y_train,
                                                   test_size = 0.2,
                                                   random_state = 17)
X_train.shape, X_valid.shape, y_train.shape, y_valid.shape
```

```python
xgb_reg = xgb.XGBClassifier().fit(X_train, y_train)
```

```python
xgb_pred = xgb_reg.predict(X_test)
xgb_pred_valid = xgb_reg.predict(X_valid)
```

```python
pred = [round(value) for value in xgb_pred]
print("Accuracy test: %.2f%%" % (accuracy_score(y_test, pred) * 100.0))

pred_valid = [round(value) for value in xgb_pred_valid]
print("Accuracy valid: %.2f%%" % (accuracy_score(y_valid, pred_valid) * 100.0))
```
